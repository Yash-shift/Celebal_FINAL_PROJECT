```python
from pyspark.sql.functions import *

CATALOG = "workspace"
GOLD_SCHEMA = "gold"
SILVER_SCHEMA = "silver"
```

```python
patients = spark.table(f"{CATALOG}.{SILVER_SCHEMA}.patients")

doctors = spark.table(f"{CATALOG}.{SILVER_SCHEMA}.doctors")

appointments = spark.table(f"{CATALOG}.{SILVER_SCHEMA}.appointments")

billing = spark.table(f"{CATALOG}.{SILVER_SCHEMA}.billing")

treatments = spark.table(f"{CATALOG}.{SILVER_SCHEMA}.treatments")
```

```python
total_patients = patients.count()

total_doctors = doctors.count()

total_appointments = appointments.count()

completed = appointments.filter(col("completed_flag")==1).count()

cancelled = appointments.filter(col("cancelled_flag")==1).count()

no_show = appointments.filter(col("no_show_flag")==1).count()

scheduled = appointments.filter(col("scheduled_flag")==1).count()

total_revenue = billing.agg(
    round(sum("successful_payment"),2)
).first()[0]

average_bill = billing.agg(
    round(avg("amount"),2)
).first()[0]
```

```python
import builtins

completion_rate = builtins.round(
    completed / total_appointments * 100,
    2
)

cancel_rate = builtins.round(
    cancelled / total_appointments * 100,
    2
)

noshow_rate = builtins.round(
    no_show / total_appointments * 100,
    2
)
```

```python
kpi = [

(
total_patients,
total_doctors,
total_appointments,
completed,
cancelled,
no_show,
scheduled,
total_revenue,
average_bill,
completion_rate,
cancel_rate,
noshow_rate
)

]

columns = [

"total_patients",

"total_doctors",

"total_appointments",

"completed",

"cancelled",

"no_show",

"scheduled",

"total_revenue",

"average_bill",

"completion_rate",

"cancel_rate",

"no_show_rate"

]

kpi_df = spark.createDataFrame(kpi,columns)

display(kpi_df)
```
